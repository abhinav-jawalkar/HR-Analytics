```python
a = ['dept','numberOfProjects','timeSpent.company','workAccident', 'promotionInLast5years','salary']
```

```python
a
```

```python
fig = plt.subplots(figsize=(20,25)) # 20,25 --> Height and width, subplots -> creating multiple plots all at once
for i,j in enumerate(a): # enumerate() helps in extracting out the index value as well as the data
    # i --> index value
    # j --> data   i = 0, j = 'dept', i = 1, j = 'numberOfProjects',...
    plt.subplot(3,2,i+1)  # 3 --> no. of rows, 2 --> columns, 
    # (3,2,1) (3,2,2) (3,2,3) (3,2,4) (3,2,5) (3,2,6)
    sns.countplot(x = j, data = df, hue = 'left') # categorical data, it will show the plots in different colors
    plt.xticks(rotation = 90)
plt.show()
```
